# The big dataset of ultra-marathon running



##### The data in this file is a large collection of ultra-marathon race records registered between 1798 and 2022 (a period of well over two centuries) being therefore a formidable long term sample. All data was obtained from public websites

##### Despite the original data being of public domain, the race records, which originally contained the athlete´s names, have been anonymized to comply with data protection laws and to preserve the athlete´s privacy. However, a column Athlete ID has been created with a numerical ID representing each unique runner (so if Antonio Fernández participated in 5 races over different years, then the corresponding race records now hold his unique Athlete ID instead of his name). This way I have preserved valuable information.
##### The dataset contains 7,461,226 ultra-marathon race records from 1,641,168 unique athletes.

```python
### make only usa based events of marathon
```

```python
df_usa = df_marathon[df_marathon['Event name'].str.contains('(USA)')]
```

```python
df_usa['Event name'].value_counts()
```

```python
### dropping the useless more null values columns 
```

```python
df_usa.drop(['Athlete club','Athlete year of birth','Athlete age category'],axis = 1,inplace=True)
```

```python
df_usa.shape
## 3 colums has been deleted
```

```python
df_usa['Event name'] = df_usa['Event name'].astype('string')
```

```python
df_usa.dtypes
```

```python
df_usa['Event name'] = df_usa['Event name'].str.replace('(USA)','',regex=False)
```

```python
#cleaning the Athlete performance

```

```python
df_usa['Athlete performance']
```

```python
df_usa['Athlete performance'] = df_marathon['Athlete performance']
### cleaning the df_usa['Athlete performance'] column
```

```python
df_usa['Athlete performance'] =df_usa['Athlete performance'].str.split(' ').str.get(0)
```

```python
### now we are taking 50km and 50mi length of usa races in 2020
```

```python
df_usa = df_usa[(df_usa['Event distance/length'].isin(['50km','50mi'])) & (df_usa['Year of event']==2020)]
```

```python
df_usa
```

```python
df_usa['Event distance/length'].value_counts()
```

```python
df_usa['Year of event'].value_counts()
```

```python
#now we are renaming the columns names for simplicity purpose
```

```python
df_usa.columns
```

```python
df_usa_2020_50mi_50km = df_usa.rename(columns = {'Year of event':'year',
                         'Event dates':'date',
                         'Event name':'event_name',
                          'Athlete ID':'a_id',
                          'Event distance/length':'distance',
                          'Event number of finishers':'noof_finishers',
                            'Athlete performance':'athlete_performance',
                         'Athlete country':'athlete_country',
                         'Athlete gender':'sex',
                         'Athlete average speed':'athelete_avg_speed'
                        
              }).reset_index(drop=True)
```

```python
df_usa_2020_50mi_50km
```

```python
df_usa_2020_50mi_50km.isna().sum()
```

```python
df_usa_2020_50mi_50km.dtypes
```

```python
df_usa_2020_50mi_50km['sex'] = df_usa_2020_50mi_50km['sex'].astype('string')
```

```python
df_usa_2020_50mi_50km['athelete_avg_speed'] = df_usa_2020_50mi_50km['athelete_avg_speed'].astype('float')
```

```python
df_usa_2020_50mi_50km['athlete_country'] = df_usa_2020_50mi_50km['athlete_country'].astype('string')
```

```python
df_usa_2020_50mi_50km['distance'] = df_usa_2020_50mi_50km['distance'].astype('string')
```

```python
df_usa_2020_50mi_50km['date'] = pd.to_datetime(df_usa_2020_50mi_50km['date'],format="%d.%m.%Y",errors='coerce')
```

```python
df_usa_2020_50mi_50km.dropna(subset = ['date'],inplace=True)
```
